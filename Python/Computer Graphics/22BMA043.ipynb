<h3>Importing the Libraries</h3>

```python
import matplotlib.pyplot as plt
```

<h3>Defining the Function to Draw the Line</h3>

```python
def line(a):
    x0 = a[0];
    y0 = a[1];
    xl = a[2];
    yl = a[3];
    dx = xl - x0;
    dy = yl - y0;
    steps = abs(dy) if abs(dy) > abs(dx) else abs(dx);
    yinc = dy / steps;
    xinc = dx / steps;
    x = [];
    y = [];
    for i in range(steps + 1):
        x.append((x0 + i * xinc));
        y.append((y0 + i * yinc));
    plt.plot(x, y);
```

<h3>Defining the Main Function</h3>

```python
def main():
    points = input("Enter the integers separated by spaces x1 y1 x2 y2 x3 y3 x4 y4(in cyclic order): ").split();
    points = [int(float(i)) for i in points]
    A = points[:4];
    line(A);
    _A = [i + 1 for i in A];
    line(_A);
    B = points[2:6];
    line(B);
    _B = [i + 1 for i in B];
    line(_B);
    C = points[4:];
    line(C);
    _C = [i + 1 for i in C];
    line(_C);
    D = points[6:] + points[:2];
    line(D);
    _D = [i + 1 for i in D];
    line(_D);
    _points = [i + 1 for i in points];
    a = points[:2] + _points[:2];
    line(a);
    b = points[2:4] + _points[2:4];
    line(b);
    c = points[4:6] + _points[4:6];
    line(c);
    d = points[6:] + _points[6:];
    line(d);
    plt.xlabel('x-axis');
    plt.ylabel('y-axis');
    plt.title('Plotting of Straight line using DDA Algorithm');
    plt.grid();
    plt.show();
```

<h3>Calling the Main Function</h3>

```python
main();
```

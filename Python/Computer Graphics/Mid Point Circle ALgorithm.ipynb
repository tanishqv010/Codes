<h2>Importing the Libraties</h2>

```python
import matplotlib.pyplot as plt
```

<h2>Taking Centre and Radius as Input</h2>

```python
x_cent = int(input("x : "))
y_cent = int(input("y : "))
r = int(input("r : "))
```

<h2>Plotting in First Octant</h2>

```python
X = [0]
Y = [r]
P = [1.25 - r]
while(X[-1] <= Y[-1]):
    if(P[-1] < 0):
        X.append(X[-1] + 1)
        Y.append(Y[-1])
        P.append(P[-1] + 2 * X[-1] + 3)
    else:
        X.append(X[-1] + 1)
        Y.append(Y[-1] - 1)
        P.append(P[-1] + 2 * X[-1] - 2 * Y[-1] + 5)
```

```python
plt.plot(X, Y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plotting of Circle using Mid Point Circle Algorithm')
plt.grid()
plt.show()
```

<h2>Plotting in First Quadant</h2>

```python
_x = [i for i in X]
_x.reverse()
_y = [i for i in Y]
_y.reverse()
X = X + _y
Y = Y + _x
```

```python
plt.plot(X, Y);
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.title('Plotting of Circle using Mid Point Circle Algorithm');
plt.grid();
plt.show();
```

<h2>Plotting in First Octant</h2>

```python
X_1 = [i for i in X];
X_2 = [-i for i in X];
X_2.reverse();
X_3 = [-i for i in X];
X_4 = [i for i in X];
X_4.reverse();
X = X_1 + X_4 + X_3 + X_2;
Y_1 = [i for i in Y];
Y_2 = [i for i in Y];
Y_2.reverse();
Y_3 = [-i for i in Y];
Y_4 = [-i for i in Y];
Y_4.reverse();
Y = Y_1 + Y_4 + Y_3 + Y_2;
```

```python
X = [x_cent + i for i in X]
Y = [y_cent + i for i in Y]
```

```python
plt.plot(X, Y);
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.title('Plotting of Circle using Mid Point Circle Algorithm');
plt.grid();
plt.show();
```

<h2>Importing the Libraries</h2>

```python
import matplotlib.pyplot as plt
```

<h2>Taking Points as Input</h2>

```python
a = input().split()
x0 = int(a[0])
y0 = int(a[1])
xl = int(a[2])
yl = int(a[3])
```

<h2>Finding Points to Draw the Line</h2>

```python
dx = abs(xl - x0)
dy = abs(yl - y0)
if(dy < dx):
    P = [2 * dy - dx]
    X = [x0]
    Y = [y0]
    while(X[-1] != xl):
        if(P[-1] < 0):
            X.append(X[-1] + 1)
            Y.append(Y[-1])
            P.append(P[-1] + 2 * dy)
        else:
            X.append(X[-1] + 1)
            Y.append(Y[-1] + 1)
            P.append(P[-1] + 2 * dy - 2 * dx)
else:
    P = [2 * dx - dy]
    X = [x0]
    Y = [y0]
    while(Y[-1] != yl):
        if(P[-1] < 0):
            X.append(X[-1])
            Y.append(Y[-1] + 1)
            P.append(P[-1] + 2 * dx)
        else:
            X.append(X[-1] + 1)
            Y.append(Y[-1] + 1)
            P.append(P[-1] + 2 * dx - 2 * dy)
```

<h2>Drawing The Line</h2>

```python
plt.plot(X, Y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plotting of Straight line using Bresenham Algorithm')
plt.grid()
plt.show()
```

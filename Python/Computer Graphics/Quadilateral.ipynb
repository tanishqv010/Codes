<h3>Importing the Libraries</h3>

```python
import matplotlib.pyplot as plt
```

<h3>Defining the Function to Draw the Line</h3>

```python
def line(a):
    x0 = a[0];
    y0 = a[1];
    xl = a[2];
    yl = a[3];
    dx = xl - x0;
    dy = yl - y0;
    steps = abs(dy) if abs(dy) > abs(dx) else abs(dx);
    yinc = dy / steps;
    xinc = dx / steps;
    x = [];
    y = [];
    for i in range(steps + 1):
        x.append((x0 + i * xinc));
        y.append((y0 + i * yinc));
    plt.plot(x, y);
```

<h3>Defining the Main Function</h3>

```python
def main():
    points = input("Enter the integers separated by spaces x1 y1 x2 y2 x3 y3 x4 y4: ").split();
    points = [int(float(i)) for i in points]
    line1 = points[:4];
    line(line1);
    line2 = points[2:6];
    line(line2);
    line3 = points[4:];
    line(line3);
    line4 = points[6:] + points[:2];
    line(line4);
    plt.xlabel('x-axis');
    plt.ylabel('y-axis');
    plt.title('Plotting of Straight line using DDA Algorithm');
    plt.grid();
    plt.show();
```

<h3>Calling the Main Function</h3>

```python
main();
```

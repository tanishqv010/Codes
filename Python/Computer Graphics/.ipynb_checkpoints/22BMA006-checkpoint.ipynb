<h3>Importing the Libraries</h3>

```python
import matplotlib.pyplot as plt
```

<h3>Defining the Function to Draw the Line</h3>

```python
def line(x0, y0, xl, yl):
    dx = abs(x0 - xl);
    dy = abs(y0 - yl);
    steps = max(dy, dx);
    yinc = dy / steps;
    xinc = dx / steps;
    x = [x0];
    y = [y0];
    for i in range(steps):
        x.append(x[-1] + xinc);
        y.append(y[-1] + yinc);
    plt.plot(x, y);
```

<h3>Defining the Main Function</h3>

```python
def main():
    x1, y1 = input("Enter 1st vertex : ").split()
    x2, y2 = input("Enter 2nd vertex : ").split()
    x3, y3 = input("Enter 3rd vertex : ").split()
    x1, y1, x2, y2, x3, y3 = int(x1), int(y1), int(x2), int(y2), int(x3), int(y3)
    line(x1, y1, x2, y2);
    line(x2, y2, x3, y3);
    line(x1, y1, x3, y3);
    plt.xlabel('X-Axis');
    plt.ylabel('Y-Axis');
    plt.title('Plotting of Straight Line using DDA Algorithm');
    plt.grid();
    plt.show();
```

<h3>Calling the Main Function</h3>

```python
main();
```

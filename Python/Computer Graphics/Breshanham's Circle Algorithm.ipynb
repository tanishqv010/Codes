<h2>Importing the Libraries</h2>

```python
import matplotlib.pyplot as plt
```

<h2>Taking Centre and Radius as Input</h2>

```python
x_cent = int(input("x : "))
y_cent = int(input("y : "))
r = int(input("r : "))
```

<h2>Plotting in First Octant</h2>

```python
x = [0]
y = [r]
p = [3 - 2 * r]
```

```python
while(x[-1] < y[-1]):
    if(p[-1] < 0):
        x.append(x[-1] + 1)
        y.append(y[-1])
        p.append(p[-1] + 6 + 4 * x[-1])
    else:
        x.append(x[-1] + 1)
        y.append(y[-1] - 1)
        p.append(p[-1] + 10 - 4 * y[-1] + 4 * x[-1])
```

```python
plt.plot(x, y);
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.title('Plotting of Circle using Breshanham\'s Circle Algorithm');
plt.grid();
plt.show();
```

<h2>Plotting in First Quadant</h2>

```python
_x = [i for i in x]
_x.reverse()
_y = [i for i in y]
_y.reverse()
X = x + _y
Y = y + _x
```

```python
plt.plot(X, Y);
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.title('Plotting of Circle using Breshanham\'s Circle Algorithm');
plt.grid();
plt.show();
```

<h2>Plotting in First Octant</h2>

```python
X_1 = [i for i in X];
X_2 = [-i for i in X];
X_2.reverse();
X_3 = [-i for i in X];
X_4 = [i for i in X];
X_4.reverse();
X = X_1 + X_4 + X_3 + X_2;
Y_1 = [i for i in Y];
Y_2 = [i for i in Y];
Y_2.reverse();
Y_3 = [-i for i in Y];
Y_4 = [-i for i in Y];
Y_4.reverse();
Y = Y_1 + Y_4 + Y_3 + Y_2;
```

```python
X = [x_cent + i for i in X]
Y = [y_cent + i for i in Y]
```

```python
plt.plot(X, Y);
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.title('Plotting of Circle using Breshanham\'s Circle Algorithm');
plt.grid();
plt.show();
```

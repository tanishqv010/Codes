<h3>Applying DDA Algorithm</h3>

<h4>Using Round Function</h4>

```python
import matplotlib.pyplot as plt
a = input().split() # Taking input in list a
x0 = int(a[0]) # Initial abscissa
y0 = int(a[1]) # Initial ordinate
xl = int(a[2]) # Final abscissa
yl = int(a[3]) # Final ordinate
dx = x0 - xl if x0 > xl else xl - x0 # Total change in abscissa
dy = y0 - yl if y0 > yl else yl - y0 # Total change in ordinate
steps = dy if dy > dx else dx # Number of steps
yinc = dy / steps # x-increment
xinc = dx / steps # y-increment
x = [] # Initializing list of abscissa of each step
y = [] # Initializing list of ordinate of each step
for i in range(steps + 1):
    x.append(round(x0 + i * xinc)) # Filling list of abscissa of each step
    y.append(round(y0 + i * yinc)) # Filling list of ordinate of each step
plt.plot(x, y) # Plotting the straight line
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plotting of Straight line using DDA Algorithm')
plt.show() # Showing the plot
```
